```python
spark = SparkSession.builder.getOrCreate()
sc = spark.sparkContext
```

```python
sqlContext = SQLContext(sc)
df = sqlContext.read.csv('/content/drive/MyDrive/Datos/TP2/user.csv', header=True, inferSchema=True)
rdd = df.rdd
```

Obtengo el usuario con mayor ratio amigos/fans:

```python
rdd.reduce(lambda x,y: x if y.fans == 0 or (x.fans != 0 and len(x.friends)/x.fans > len(y.friends)/y.fans) else y)
```

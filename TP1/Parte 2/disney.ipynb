### Disney movies

Dataset: [Disney Movies](https://www.kaggle.com/datasets/prateekmaj21/disney-movies?select=disney_movies.csv)

Target variable: `inflation_adjusted_gross`

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

disney_movies = pd.read_csv('/content/drive/MyDrive/Datos/TP1/Parte 2/disney_movies.csv')
disney_movies
```

Ajusto la unidad de la columna inflation_adjusted_gross a millones de dÃ³lares, para poder realizar los plots en esa unidad sin usar notaciÃ³n cientÃ­fica.

```python
disney_movies['inflation_adjusted_gross'] = disney_movies['inflation_adjusted_gross'].map(lambda x: x/10e5)
disney_movies
```

```python
disney_movies['mpaa_rating'].value_counts()
```

CalificaciÃ³n MPAA:

- **G â€“ General Audiences**. Admitido para todas las edades. Nada que pudiera ofender a los padres si lo vieran sus hijos.

- **PG â€“ Parental Guidance Suggested**. Algunos contenidos pueden no ser apropiados para niÃ±os. Puede contener algunas imÃ¡genes que podrÃ­an no gustar a los padres para sus hijos jÃ³venes.

- **PG-13 â€“ Parents Strongly Cautioned**. Algunos materiales pueden ser inapropiados para niÃ±os menores de 13 aÃ±os. Los padres deben ser cautelosos. Puede ser inapropiado para pre-adolescentes.

- **R â€“ Restricted**. Personas de menos de 17 aÃ±os requieren acompaÃ±amiento de un adulto. Contienen algunos materiales para adultos. Los padres deben informarse sobre la pelÃ­cula antes de llevar a sus hijos a verla.

```python
plt.figure(dpi=150)
sns.barplot(data=disney_movies, x="mpaa_rating", y="inflation_adjusted_gross", ci=None, 
            order=['G', 'PG', 'PG-13', 'R'], palette=['#1baaf7', '#f7f71b', '#f7861b', '#f72a1b'])
plt.title("RecaudaciÃ³n promedio de las pelÃ­culas de Disney\n segÃºn su clasificaciÃ³n MPAA")
plt.xlabel("ClasificaciÃ³n MPAA")
plt.ylabel("RecaudaciÃ³n (en millones de dÃ³lares)")
plt.show()
```

```python
plt.figure(dpi=150)
g = sns.boxplot(data=disney_movies, y='genre', x='inflation_adjusted_gross', order=['Action', 'Adventure', 'Comedy', 'Drama'], palette=['#e0130b', '#e66400', '#eded00', '#00cee6'])
g.set_yticklabels(["AcciÃ³n", "Aventura", "Comedia", "Drama"])
plt.xlim(-0.0005,0.8e3)
plt.ticklabel_format(style='plain', axis='x')
plt.title("RecaudaciÃ³n de las pelÃ­culas de Disney\nsegÃºn su gÃ©nero")
plt.ylabel("GÃ©nero")
plt.xlabel("RecaudaciÃ³n (en millones de dÃ³lares)")
plt.show()
```

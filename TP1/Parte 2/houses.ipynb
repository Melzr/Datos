### House prices

Dataset: [House Prices](https://www.kaggle.com/competitions/house-prices-advanced-regression-techniques/data?select=train.csv)

Target variable: `SalePrice`

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

houses = pd.read_csv('/content/drive/MyDrive/Datos/TP1/Parte 2/houses.csv')
houses
```

```python
houses.columns
```

```python
plt.figure(dpi=150)
plt.hist(houses.SalePrice, color='#38e043')
plt.title("Distribution of houses sell price")
plt.xlabel("Sell price in usd")
plt.ylabel("Number of houses sold")
plt.show()
```

```python
houses.LandContour.value_counts()
```

```python
plt.figure(dpi=150)
g = sns.violinplot(data=houses, x='LandContour', y='SalePrice', order=['Low', 'Lvl', 'Bnk', 'HLS'], palette=['#d6410b', '#ffec1c', '#52d91c', '#17850b'])
g.set_xticklabels(['Depression', 'Near Flat/Level', 'Banked', 'Hillside'])
plt.xlabel('Land contour')
plt.ylabel('Sale price in usd')
plt.title('Houses sell price according to land contour')
plt.show()
```

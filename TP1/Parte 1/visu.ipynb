```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

user_friends = pd.read_csv('/content/drive/MyDrive/Datos/TP1/Parte 1/user.csv', usecols=['friends'])
user_friends
```

Calculo la cantidad de amigos que tiene cada usuario:

```python
def number_of_friends(friends):
  if friends == "None":
    return 0 
  return len(friends.split(','))

user_friends['number_of_friends'] = user_friends['friends'].map(number_of_friends)
user_friends
```

```python
user_friends.number_of_friends.value_counts()
```

Para obtener el mismo gr√°fico, me quedo solo con los usuarios que tienen entre 0 y 600 amigos:

```python
user_friends = user_friends[user_friends['number_of_friends'] <= 600]
percentil_80 = user_friends['number_of_friends'].quantile(0.8)
```

Grafico la densidad de la cantidad de amigos y el cuantil 80 de la misma:

```python
plt.figure(dpi=150)
sns.kdeplot(user_friends['number_of_friends'], color='#B7EA53', shade=True)
plt.xlim(0, 600)
plt.yticks([])
plt.title('Ley de potencias para la cantidad de amigos')
plt.ylabel('')
plt.xlabel('Cantidad de amigos')
plt.axvline(x=percentil_80, color='#DEA82A', label='Percentil 80', linestyle='--')
plt.legend()
plt.show()
```
